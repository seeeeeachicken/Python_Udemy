```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
d_range = pd.date_range(start='2020/1/1', end='2020/1/8')
names = list('ABCD')
temp = {'value':[], 'variable':[], 'date': []}

for name in names:
    for day in d_range:
        temp['value'].append(np.random.random())
        temp['variable'].append(name)
        temp['date'].append(day)
        
dframe = pd.DataFrame(temp)
```

```python
dframe
```

```python
dframe_piv = dframe.pivot(index='date', columns='variable', values='value')
dframe_piv
```
